```python
# import Regex library for string manipulation 
import re, string, unicodedata
```

```python
# a function to clean tweets and remove special characters and emoji from tweet
def clean(text):
    text = re.sub('[^A-Za-z]+',' ',text)
    return text
```

```python
# apply the clean function to tweet and store clean text into a new column
df['tweet_clean'] = df['Tweet'].apply(clean)
```

```python
#import text preprocessing libraries 

import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag
from nltk.corpus import stopwords
from nltk.corpus import wordnet

nltk.download('wordnet')

nltk.download('punkt')

nltk.download('averaged_perceptron_tagger')

nltk.download('stopwords')
```

```python
#create a dictionary for parts of speech 
pos_dict = {'J':wordnet.ADJ, 'V' : wordnet.VERB, 'N' : wordnet.NOUN, 'R' :wordnet.ADV}

#define a function for Part of speech, Stopwords and Tokenization
def token_stop_pos(text):
    tags = pos_tag(word_tokenize(text))
    list = []
    for line, tag in tags:
        if line.lower() not in set(stopwords.words('english')):
            list.append(tuple([line,pos_dict.get(tag[0])]))
    return list
```

```python
df.head()
```

```python
#create new column to store tokenization
df['pro_twitter'] = df['tweet_clean'].apply(token_stop_pos)
```

```python
df.head()
```

```python
#import lemmatization library and define a function for the process
from nltk.stem import WordNetLemmatizer

word_net_lemmatizer = WordNetLemmatizer()

#lemmatizer process function
def lemmatize(pos_data):
        lemma_rew = ""
        for word, pos in pos_data:
            if not pos:
                lemma = word
                lemma_rew = lemma_rew
            else:
                lemma = word_net_lemmatizer.lemmatize(word,pos=pos)
                lemma_rew =lemma_rew+" "+lemma
        return lemma_rew
```

```python
df['lemmatized'] = df['pro_twitter'].apply(lemmatize)
```

```python
df.head()
```

```python
#import textblob library for Sentiment analysis and define function

from textblob import TextBlob

def get_sentiment(text):
    analysis = TextBlob(text)
    if analysis.sentiment.polarity > 0:
        return 'Positive'
    elif analysis.sentiment.polarity == 0:
        return 'Neutral'
    else:
        return 'Negative'
```

```python
df['tblob_Sentiment'] = df['lemmatized'].apply(get_sentiment)
```

```python
df.head()
```

```python
#import vader library for Sentiment analysis and define function

!pip install vaderSentiment

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()

def vader_analysis(review):
        vs = analyzer.polarity_scores(review)
        return vs ['compound']

df['vader_comp'] = df['lemmatized'].apply(vader_analysis)


def vader_sentiment(comp):
        if comp > 0:
            return 'Positive'
        elif comp < 0:
            return 'Negative'
        else:
            return 'Neutral'
        
```

```python
df['vader_Sentiment'] = df['vader_comp'].apply(vader_sentiment)
```
